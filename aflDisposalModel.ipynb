```python
df_team_stats = df_stats.groupby(['gameId','team'], as_index=False).sum().drop(['playerId', '% Played', 'Brownlow Votes', 'gameNumber'],1)

for i, val in enumerate(df_team_stats['gameId']):
    df_team_stats['year'][i] = int(val[0:4])


df_team_stats['Score'] = df_team_stats.apply(lambda row: row['Goals']*6 + row['Behinds'], axis=1)
    
df_team_stats
```

```python
df = pd.merge(df_team_stats, df_games, suffixes=('', '_y'), on = 'gameId')

df.drop(df.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)

df.drop(['date', 'attendance','homeTeamScore','awayTeamScore','homeTeam','awayTeam','round'],1)
```

```python
df_games['date'] = pd.to_datetime(df_games['date'])
df_games = df_games.sort_values(by="date")

current_elo = dict((team,1500) for team in df_games.homeTeam.unique())

elo_tracker = pd.DataFrame(current_elo.items(),columns = ['team','elo'])
elo_tracker['date'],elo_tracker['round'] = [pd.to_datetime('01-Jan-2012'),'R0']

pred_tracker = []

for index, row in df_games.iterrows():
    elo_home = current_elo[row.homeTeam]
    elo_away = current_elo[row.awayTeam]

    mov = abs(row.homeTeamScore - row.awayTeamScore)

    exp_home = 10**(elo_home/400)/(10**(elo_home/400) + 10**(elo_away/400))
    exp_away = 1-exp_home


    if (row.homeTeamScore > row.awayTeamScore):
        elo_diff = elo_home - elo_away
        k = 20*((mov + 3)**0.8)/(7.5+0.006*elo_diff)

        current_elo[row.homeTeam] = k*(1-exp_home) + current_elo[row.homeTeam]
        current_elo[row.awayTeam] = k*(0-exp_away) + current_elo[row.awayTeam]

        if (elo_home > elo_away):
            pred_tracker.append(1)

        else: 
            pred_tracker.append(0)

    elif (row.homeTeamScore < row.awayTeamScore):
        elo_diff = elo_away - elo_home
        k = 20*((mov + 3)**0.8)/(7.5+0.006*elo_diff)

        current_elo[row.homeTeam] = k*(0-exp_home) + current_elo[row.homeTeam]
        current_elo[row.awayTeam] = k*(1-exp_away) + current_elo[row.awayTeam]

        if (elo_home < elo_away):
            pred_tracker.append(1)

        else: 
            pred_tracker.append(0)

    else:
        continue

    elo_tracker = elo_tracker.append({'team':row.homeTeam,'elo':current_elo[row.homeTeam], 'date':row.date,'round':row['round']},ignore_index=True)
    elo_tracker = elo_tracker.append({'team':row.awayTeam,'elo':current_elo[row.awayTeam], 'date':row.date,'round':row['round']},ignore_index=True)

    if row['round'] == 'GF':
        for team in df_games.homeTeam.unique():
            current_elo[team] = current_elo[team]*0.75 + 0.25*1505
            elo_tracker = elo_tracker.append({'team':team,'elo':current_elo[team], 'date':pd.to_datetime('01-Jan-' + str(row.year+1)),'round':'R0'},ignore_index=True)


current_elo
```

```python
fig = px.line(elo_tracker, x="date", y="elo", color='team',color_discrete_sequence = ['#f15c22','#fed102','#4d2004','#013b9f', '#cc2031','#d93e39','#2a0d54','#013b9f','#008aab','#a30046','#000000','#ffd200','#ed171f','#cc2031','#062ee2','#ed0f05','#1c3c63','#0e1e2d'])
fig.show()
```

```python
import numpy as np

pred_cumsum = np.cumsum(pred_tracker)
pred_cumsum
pred_ave = []

for n,item in enumerate(pred_cumsum):
    pred_ave.append(item/(n+1))

fig = px.line(pred_ave)
fig.show()
```
